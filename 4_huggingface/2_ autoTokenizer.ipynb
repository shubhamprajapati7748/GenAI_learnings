## Tokenizer
A tokenizer converts your input into a format that can be processed by the model.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

sentence -> Token -> Embedding -> Position encoding -> self_attention -> feedForward -> o/p

```python
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
# model = AutoModelForSequenceClassification.from_pretrained("google/flan-t5-large")
```

```python
save_directory ="saved_models"
```

## saving the model and tokenizer

```python

tokenizer.save_pretrained(save_directory=save_directory)
```
